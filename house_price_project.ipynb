```python
import pandas as pd
import pandas as pd
# To supress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('D://Pandas/house_prices.csv')
df.head(3)
```

```python
df.drop(['Index','Description','Society','Car Parking','Super Area','Dimensions','Plot Area'],axis=1,inplace=True)
```

```python
df['Title']=df['Title'].str[:5]
```

```python
# converting lac nd cr in amount
def convert_amount(val):
    val = val.strip()
    if 'Lac' in val:
        return float(val.replace('Lac', '').strip()) * 1e5
    elif 'Cr' in val:
        return float(val.replace('Cr', '').strip()) * 1e7
    else:
        return None
```

```python
df['Amount(in rupees)'] = df['Amount(in rupees)'].apply(convert_amount)
```

```python
# Convert 'Amount(in rupees)' to numeric and fill nullvalues
#df['Amount(in rupees)'].fillna(df['Amount(in rupees)'].mean(),inplace=True)
df['Amount(in rupees)'] = pd.to_numeric(df['Amount(in rupees)'], errors='coerce')

# Convert 'Price (in rupees)' to numeric, then fill missing with mean
df['Price (in rupees)'] = pd.to_numeric(df['Price (in rupees)'], errors='coerce')
df['Price (in rupees)'].fillna(df['Price (in rupees)'].mean(), inplace=True)

# Clean 'Carpet Area' from 'sqft' and convert to numeric
df['Carpet Area'] = df['Carpet Area'].str.replace('sqft', '', case=False).str.strip()
df['Carpet Area'] = pd.to_numeric(df['Carpet Area'], errors='coerce')
df['Carpet Area'].fillna(df['Carpet Area'].median(), inplace=True)

# Extract floor number (first 2 characters) and convert
df['Floor'] = df['Floor'].str[:2]
df['Floor'] = pd.to_numeric(df['Floor'], errors='coerce')
df['Floor'].fillna(df['Floor'].median(), inplace=True)

# Fill missing 'Balcony' with median and convert to numeric
df['Balcony'] = pd.to_numeric(df['Balcony'], errors='coerce')
df['Balcony'].fillna(df['Balcony'].mode()[0], inplace=True)

# fill missing 'status' with mode 
df['Status'].fillna(df['Status'].mode()[0],inplace=True)

# fill missing 'Transaction' with mode
df['Transaction'].fillna(df['Transaction'].mode()[0],inplace=True)

# fill missing 'Furnishing' with mode
df['Furnishing'].fillna(df['Furnishing'].mode()[0],inplace=True)

# fill missing 'Facing' with mode
df['facing'].fillna(df['facing'].mode()[0],inplace=True)

# fill missing 'overlooking' with mode
df['overlooking'].fillna(df['overlooking'].mode()[0],inplace=True)

# fill missing 'Bathroom' and convert to numeric 
df['Bathroom'].fillna(df['Bathroom'].mode()[0],inplace=True)
df['Bathroom']=pd.to_numeric(df['Bathroom'], errors='coerce')

# fill missing 'Owenrship' with mode
df['Ownership'].fillna(df['Ownership'].mode()[0],inplace=True)

df['Bathroom'].fillna(df['Bathroom'].mode()[0],inplace=True)
```

```python
x=df[['Title', 'Price (in rupees)', 'location',
       'Carpet Area', 'Status', 'Floor', 'Transaction', 'Furnishing', 'facing',
       'overlooking', 'Bathroom', 'Balcony', 'Ownership']]
y=df['Amount(in rupees)']
```

```python
cat_x=x.select_dtypes(exclude='number')
num_x=x.select_dtypes(include='number')
```

```python
cat_x=pd.get_dummies(cat_x)
cat_x.head()
```

```python
x=pd.concat([cat_x,num_x],axis=1)
x.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(x,y,test_size=0.1,random_state=7)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
y_pred=model.predict(X_test)
y_actual=y_test
```
